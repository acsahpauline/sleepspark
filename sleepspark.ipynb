```python
import zipfile

def unzip_to_subfolder(zip_path, extract_to):
    dataset_name = os.path.splitext(os.path.basename(zip_path))[0]
    extract_path = os.path.join(extract_to, dataset_name)
    os.makedirs(extract_path, exist_ok=True)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"Unzipped {zip_path}  {extract_path}")
for item in os.listdir(base_path):
    if item.endswith('.zip'):
        zip_path = os.path.join(base_path, item)
        unzip_to_subfolder(zip_path, base_path)

print("\nAfter unzipping, top-level contents:")
!ls -l {base_path}
```

```python
import zipfile
import os

def unzip_specific(zip_name, base_path):
    zip_path = os.path.join(base_path, zip_name)
    if not os.path.exists(zip_path):
        print(f"Zip not found: {zip_path}")
        return

    dataset_name = os.path.splitext(zip_name)[0]
    extract_path = os.path.join(base_path, dataset_name)
    os.makedirs(extract_path, exist_ok=True)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"Unzipped {zip_name}  {extract_path}")

unzip_specific('us-accidents.zip', base_path)
unzip_specific('sleep-health-and-lifestyle-dataset.zip', base_path)
unzip_specific('API_NY.GDP.MKTP.CD_DS2_en_csv_v2_280632.zip', base_path)

print("\nus-accidents folder:")
!ls -lh {base_path}/us-accidents
```

```python
!pip install pyspark

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Sleep Deprivation - Large Scale Processing") \
    .config("spark.executor.memory", "4g") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

print("Spark version:", spark.version)
spark.sparkContext.setLogLevel("ERROR")
```

```python
accidents_path = "/content/datasets/us-accidents/US_Accidents_March23.csv"
df_acc = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(accidents_path)

print("US Accidents loaded successfully!")
print("Number of rows:", df_acc.count())
print("Number of columns:", len(df_acc.columns))
df_acc.printSchema()
df_acc.show(5, truncate=False)
```

```python
from pyspark.sql.functions import col
print("Default partitions:", df_acc.rdd.getNumPartitions())
df_acc = df_acc.repartition(200)
print("After repartition(200):", df_acc.rdd.getNumPartitions())
df_acc.cache()
df_acc.count()
print("DataFrame cached.")
print("\nPhysical plan example:")
df_acc.filter(col("Severity") == 3).explain()
```

```python
from pyspark.sql.functions import count, when, col, sum as spark_sum
null_counts = df_acc.select(
    [spark_sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in ["Start_Time", "Severity", "City", "State"]]
).collect()[0]
print("Null counts in key columns:")
for col_name, cnt in null_counts.asDict().items():
    print(f"  {col_name}: {cnt:,}")
dup_count = df_acc.groupBy("ID").count().filter(col("count") > 1).count()
print(f"Duplicate IDs: {dup_count:,}")
```

```python
from pyspark.sql.functions import count
state_counts = df_acc.groupBy("State").agg(count("*").alias("Accident_Count")) \
    .orderBy(col("Accident_Count").desc())
print("Top 10 states by accident count:")
state_counts.show(10)
state_counts.cache()
```

```python
from pyspark.sql.functions import (
    hour, minute, dayofweek, month, col, when, lit, to_date,
    date_format, lag, sum as spark_sum, broadcast
)
from pyspark.sql.window import Window
print("Starting feature engineering on 7.7M rows...")
df_acc_enriched = df_acc.withColumn("Hour_of_Day", hour("Start_Time")) \
    .withColumn("Minute", minute("Start_Time")) \
    .withColumn("Day_of_Week", dayofweek("Start_Time")) \
    .withColumn("Month", month("Start_Time")) \
    .withColumn("Date", to_date("Start_Time")) \
    .withColumn("Is_Night", col("Hour_of_Day").between(0, 5) | (col("Hour_of_Day") == 23)) \
    .withColumn("Is_Drowsy_Window", col("Hour_of_Day").between(0, 5)) \
    .withColumn("Season",
        when(col("Month").isin(12, 1, 2), "Winter")
        .when(col("Month").isin(3, 4, 5), "Spring")
        .when(col("Month").isin(6, 7, 8), "Summer")
        .otherwise("Fall")
    )
df_acc_enriched.cache()
print("Enriched DataFrame columns:")
print(df_acc_enriched.columns)
df_acc_enriched.select(
    "Start_Time", "Hour_of_Day", "Is_Drowsy_Window", "Season", "Severity", "State"
).show(10, truncate=False)
```

```python
window_spec = Window.partitionBy("State").orderBy("Date").rowsBetween(-6, 0)

df_with_rolling = df_acc_enriched.withColumn(
    "Rolling_7Day_Accidents",
    spark_sum(lit(1)).over(window_spec)
)

print("Sample with rolling 7-day count (one state):")
df_with_rolling.filter(col("State") == "CA").select(
    "Date", "State", "Rolling_7Day_Accidents"
).orderBy("Date").show(10, truncate=False)

df_with_rolling.cache()
```

```python
from pyspark.sql.functions import count, col, when, round
drowsy_stats = df_acc_enriched.groupBy("Is_Drowsy_Window").agg(
    count("*").alias("Accident_Count"),
    round((count("*") / df_acc_enriched.count()) * 100, 2).alias("Percentage")
).orderBy("Is_Drowsy_Window")

print("Accidents in Drowsy Window (00:00-05:59) vs Rest of Day:")
drowsy_stats.show()
severity_by_drowsy = df_acc_enriched.groupBy("Is_Drowsy_Window").agg(
    round(avg("Severity"), 2).alias("Avg_Severity")
)

print("Average Severity in Drowsy vs Non-Drowsy:")
severity_by_drowsy.show()
```

```python
from pyspark.ml.stat import ChiSquareTest
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors

df_test = df_acc_enriched.withColumn("Severity_Bin",
    when(col("Severity") >= 3, "High").otherwise("Low")
).select("Is_Drowsy_Window", "Severity_Bin")

assembler = VectorAssembler(inputCols=["Severity_Bin_index"], outputCol="features")  # need indexing

contingency = df_test.groupBy("Is_Drowsy_Window", "Severity_Bin").count()
contingency.show()

print("Contingency table for chi-square test (High vs Low severity by drowsy window)")
```

```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml.stat import ChiSquareTest
from pyspark.ml.linalg import Vectors
from pyspark.sql.functions import col

indexer = StringIndexer(inputCol="Severity_Bin", outputCol="Severity_Index")
df_indexed = indexer.fit(df_test).transform(df_test)

contingency_rdd = df_indexed.groupBy("Is_Drowsy_Window", "Severity_Index").count() \
    .rdd.map(lambda row: (row["Is_Drowsy_Window"], row["Severity_Index"], row["count"]))

import pandas as pd
from scipy.stats import chi2_contingency
cont_df = contingency.toPandas().pivot(index="Is_Drowsy_Window", columns="Severity_Bin", values="count").fillna(0)

print("Contingency table (pandas):")
print(cont_df)
chi2, p, dof, expected = chi2_contingency(cont_df.values)

print(f"\nChi-square statistic: {chi2:.2f}")
print(f"p-value: {p:.2e}")
print("Significant difference (p < 0.05):", p < 0.05)
```

```python
from pyspark.sql.functions import col, lit

sleep_main_path = "/content/datasets/sleep-health-and-lifestyle-dataset/Sleep_health_and_lifestyle_dataset.csv"
df_main = spark.read.option("header", "true").option("inferSchema", "true").csv(sleep_main_path)
df_main = df_main.withColumn("Source", lit("Main_Lifestyle"))

sleep_eff_path = "/content/datasets/sleep-efficiency/Sleep_Efficiency.csv"
df_eff = spark.read.option("header", "true").option("inferSchema", "true").csv(sleep_eff_path)
df_eff = df_eff.withColumn("Source", lit("Sleep_Efficiency"))

health_sleep_path = "/content/datasets/health-and-sleep-statistics/health_and_sleep_statistics.csv"
try:
    df_health = spark.read.option("header", "true").option("inferSchema", "true").csv(health_sleep_path)
    df_health = df_health.withColumn("Source", lit("Health_Stats"))
except Exception as e:
    print("Health stats not loaded:", str(e))
    df_health = None

print("Main DF schema:")
df_main.printSchema()

print("\nEfficiency DF schema:")
df_eff.printSchema()

common_cols = list(set(df_main.columns) & set(df_eff.columns))
print("\nCommon columns for union:", common_cols)

df_union = df_main.select(common_cols).union(df_eff.select(common_cols)) if common_cols else df_main

print("\nUnioned sleep data rows:", df_union.count())
df_union.show(5, truncate=False)
```

```python
print("Sleep Duration by Stress Level (Main Dataset):")
df_main.groupBy("Stress Level").agg(
    round(avg("Sleep Duration"), 2).alias("Avg_Sleep_Hours"),
    count("*").alias("Count")
).orderBy("Stress Level").show()

print("\nSleep Quality by BMI Category:")
df_main.groupBy("BMI Category").agg(
    round(avg("Quality of Sleep"), 2).alias("Avg_Quality"),
    count("*").alias("Count")
).show()

print("\nHigh-risk % by Physical Activity Level (binned):")
df_main.withColumn("Activity_Bin",
    when(col("Physical Activity Level") < 50, "Low")
    .when(col("Physical Activity Level") < 80, "Medium")
    .otherwise("High")
).groupBy("Activity_Bin").agg(
    round((count(when(col("Sleep Duration") < 7, True)) / count("*")) * 100, 2).alias("Deprived_%")
).show()
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.sql.functions import col, when

df_ml = df_main.withColumn("High_Risk",
    when(
        (col("Sleep Duration") < 7) |
        (col("Quality of Sleep") < 6) |
        (col("Sleep Disorder") != "None"), 1
    ).otherwise(0)
)

indexers = [
    StringIndexer(inputCol="Occupation", outputCol="Occupation_Index", handleInvalid="keep"),
    StringIndexer(inputCol="Gender", outputCol="Gender_Index", handleInvalid="keep"),
    StringIndexer(inputCol="BMI Category", outputCol="BMI_Index", handleInvalid="keep")
]

assembler = VectorAssembler(
    inputCols=["Age", "Stress Level", "Physical Activity Level",
               "Occupation_Index", "Gender_Index", "BMI_Index"],
    outputCol="features",
    handleInvalid="skip"
)

pipeline = Pipeline(stages=indexers + [assembler])

pipeline_model = pipeline.fit(df_ml)
df_features = pipeline_model.transform(df_ml)

train_data, test_data = df_features.randomSplit([0.8, 0.2], seed=42)

lr = LogisticRegression(labelCol="High_Risk", featuresCol="features", maxIter=10)
lr_model = lr.fit(train_data)

predictions = lr_model.transform(test_data)

import builtins

evaluator_auc = BinaryClassificationEvaluator(labelCol="High_Risk", metricName="areaUnderROC")
auc = evaluator_auc.evaluate(predictions)

evaluator_acc = MulticlassClassificationEvaluator(labelCol="High_Risk", metricName="accuracy")
accuracy = evaluator_acc.evaluate(predictions)

print(f"Model Performance on Test Set:")
print(f" - AUC (ROC): {builtins.round(auc, 3)}")
print(f" - Accuracy: {builtins.round(accuracy * 100, 1)}%")
print("\nSample Predictions:")
predictions.select(
    "Age", "Stress Level", "Occupation", "Gender", "High_Risk", "prediction", "probability"
).show(10, truncate=False)

print("\nFeature Coefficients (importance):")
coefs = lr_model.coefficients.toArray()
feature_names = ["Age", "Stress Level", "Physical Activity Level", "Occupation_Index", "Gender_Index", "BMI_Index"]

for name, coef in zip(feature_names, coefs):
    print(f"{name}: {builtins.round(float(coef), 4)}")
```

```python
import pandas as pd
import plotly.express as px
from scipy.stats import linregress
import numpy as np

import builtins
round = builtins.round


trend_data = pd.DataFrame({
    "Year": [2000, 2005, 2010, 2015, 2016, 2018, 2022, 2025],
    "Percent_Deprived": [30, 32, 35, 37, 35, 40, 38, 60],
    "Source": ["Historical"] * 8
})

your_data = pd.DataFrame({
    "Year": [2025],
    "Percent_Deprived": [62.1],
    "Source": ["Your Teen Dataset"]
})

slope_lin, intercept_lin, r_lin, p_lin, std_lin = linregress(trend_data["Year"], trend_data["Percent_Deprived"])
proj_2030_lin = slope_lin * 2030 + intercept_lin
proj_2035_lin = slope_lin * 2035 + intercept_lin

log_y = np.log(trend_data["Percent_Deprived"])
slope_exp, intercept_exp, r_exp, p_exp, std_exp = linregress(trend_data["Year"], log_y)
proj_2030_exp = np.exp(slope_exp * 2030 + intercept_exp)
proj_2035_exp = np.exp(slope_exp * 2035 + intercept_exp)

print("Linear Projection:")
print(f"2030: {round(proj_2030_lin, 1)}%")
print(f"2035: {round(proj_2035_lin, 1)}%")
print(f"R-squared: {round(r_lin**2, 3)}, p-value: {p_lin:.2e}")

print("\nExponential Projection (if trend accelerates):")
print(f"2030: {round(proj_2030_exp, 1)}%")
print(f"2035: {round(proj_2035_exp, 1)}%")
print(f"R-squared: {round(r_exp**2, 3)}, p-value: {p_exp:.2e}")

all_data = pd.concat([
    trend_data,
    your_data,
    pd.DataFrame({
        "Year": [2030, 2035],
        "Percent_Deprived": [proj_2030_lin, proj_2035_lin],
        "Source": ["Linear Projection", "Linear Projection"]
    })
], ignore_index=True)

fig = px.line(
    all_data,
    x="Year",
    y="Percent_Deprived",
    color="Source",
    title="Sleep Deprivation Trend & Projection (% Adults Not Getting 7+ Hours)",
    labels={"Percent_Deprived": "% Adults", "Year": "Year"},
    markers=True,
    range_y=[0, 100]
)

fig.add_annotation(
    x=2025, y=62.1,
    text="Teen Dataset: 62.1% deprived",
    showarrow=True,
    arrowhead=1,
    ax=40,
    ay=-40
)

fig.update_layout(
    hovermode="x unified",
    height=600
)

fig.show()
```
